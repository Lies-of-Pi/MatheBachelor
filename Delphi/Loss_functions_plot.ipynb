# Plots zu den Verlustfunktionen

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# L2 / Exp-Dist Loss with Normal distribution

# true data
mu_true, sigma = 4, 1 # mean and standard deviation
y = np.random.normal(mu_true, sigma, 10000)

# mu parameters for the model
mu_values = np.linspace(1, 7, 1000)

# calculate Losses:
l2_losses = []
exp_dist_losses = []
for mu in mu_values:
    y_prediction = np.random.normal(mu, sigma, 10000)
    l2_losses.append( np.mean( (y - y_prediction)**2 ) )
    exp_dist_losses.append(- np.mean(np.log(mu) - mu * y))

# Plot
plt.figure(dpi = 120)
plt.plot(mu_values, l2_losses, label = "L2 Loss")
plt.plot(mu_values, exp_dist_losses, label="Exp-Dist Loss")
plt.axvline(mu_true, label="True parameter", color="red")
plt.grid()
plt.legend()
plt.xlabel("Model parameter")
plt.ylabel("Loss value")
plt.title("Normal distribution")
plt.savefig("Normal_Dist_Plot.png")
plt.show()
```

```python
# L2 / Exp-Dist Loss with Exponential distribution

# true data
lambda_true = 4
y = np.random.exponential(scale = 1/lambda_true, size=10000) #### die Numpy Exponentialfunktion nimmt den Kehrwert von Lambda als Argument!!!!

# lambda parameters for the model
lambda_values = np.linspace(1, 7, 1000)

# calculate Losses:
l2_losses = []
exp_dist_losses = []
for lamb in lambda_values:
    y_prediction = np.random.exponential(scale = 1/lamb, size=10000) #### die Numpy Exponentialfunktion nimmt den Kehrwert von Lambda als Argument!!!!
    l2_losses.append( np.mean( (y - y_prediction)**2 ) )
    exp_dist_losses.append(- np.mean(np.log(lamb) - lamb * y))

# Plot
plt.figure(dpi = 120)
plt.plot(lambda_values, l2_losses, label = "L2 Loss")
plt.plot(lambda_values, exp_dist_losses, label="Exp-Dist Loss")
plt.axvline(lambda_true, label="True parameter", color="red")
plt.grid()
plt.legend()
plt.xlabel("Model parameter")
plt.ylabel("Loss value")
plt.title("Exponential distribution")
plt.savefig("Exp_Dist_Plot.png")
plt.show()
```
